

# **Color Spaces**

In this lesson we'll learn to:
1. View the individual channels of an RGB Image
2. Manipulate a color space
3. Introduce HSV Color Spaces

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

def imshow(image, title="", size=10):
    w, h= image.shape[0], image.shape[1]
    aspect_ratio= w/h
    plt.figure(figsize=(size*aspect_ratio, size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
```

```python
image=cv2.imread("images/castara.jpg")

B, G, R= cv2.split(image)

print(B.shape)
print(G.shape)
print(R.shape)
```

```python
imshow(image, "castara")
```

```python
imshow(B, "Blue channel only")
```

```python
image.dtype, B.dtype
```

```python
zeros= np.zeros(image.shape[:2], dtype="uint8")

zeros.shape, image.shape[:2]
```

```python
cv2.merge([zeros, zeros, B]).shape
```

```python
imshow(cv2.merge([zeros, zeros, R]), 'Red channel')
imshow(cv2.merge([zeros, G, zeros]), 'Green channel')
imshow(cv2.merge([B, zeros, zeros]), 'Blue channel')
```

```python
merged= cv2.merge([B,G,R])
imshow(merged, "Merged")
```

```python
merged= cv2.merge([B+100, G, R])
imshow(merged)
```

## **The HSV Color Space**

![](https://upload.wikimedia.org/wikipedia/commons/f/f2/HSV_color_solid_cone.png)

- Hue: 0 - 179 
- Saturation: 0 - 255
- Value (Intensity): 0 - 255

```python
image= cv2.imread('images/castara.jpg')

# Convert to HSV
hsv_image= cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
imshow(hsv_image, 'HSV')
```

```python
hsv_image.shape
```

This looks odd....why? 

Because our plotting function was designed for RGB only images, not HSV

```python
plt.imshow(cv2.cvtColor(hsv_image, cv2.COLOR_HSV2RGB))
```

### **Let's view each channel type in the HSV Color Space representation**

```python
imshow(hsv_image[:,:,0], 'Hue')
imshow(hsv_image[:,:,1], 'Saturation')
imshow(hsv_image[:,:,2], 'Value')
```
