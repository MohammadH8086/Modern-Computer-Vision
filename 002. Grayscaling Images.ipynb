# Grayscaling Images
In this lesson we'll learn to:

- Convert a color image to grayscale
- See the change in dimensions between grayscale and color images

```python
import cv2
from matplotlib import pyplot as plt
```

```python
def imshow(image=None, size=10, title=""):
    w, h= image.shape[0], image.shape[1]
    aspect_ratio= w/h
    plt.figure(figsize=(size*aspect_ratio, size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(f'{title},   aspect ratio: {aspect_ratio:.2f},   shape: {image.shape}')
    
    
```

```python
image=cv2.imread("images/castara.jpg")
imshow(image,  title="Castara")
```

```python
def imshow(image=None, size=10, title=""):
    # Change from w,h to h,w
    h, w= image.shape[0], image.shape[1]
    aspect_ratio= w/h
    plt.figure(figsize=(size*aspect_ratio, size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(f'{title},   aspect ratio: {aspect_ratio:.2f},   shape: {image.shape}')
    
```

```python
imshow(image,  title="Castara")
```

```python
gray_image= cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
imshow(gray_image, size=4, title="Converted to Grayscale")
```

### **Grayscale Image Dimensions**

Remember RGB color images have 3 dimensions, one for each primary color. Grayscale just has 1, which is the intensity of gray.
