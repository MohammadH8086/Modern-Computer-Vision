
## **Drawing images and shapes using OpenCV**



````python
# Our Setup, Import Libaries, Create our Imshow Function and Download our Images
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Define our imshow function 
def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
````

````python
# Create a black image
image= np.zeros((512,512,3), np.uint8)

# Create a grayscale image?
image_gray= np.zeros((512,512), np.uint8)

imshow('Black Canvas - RGB Color', image, 4)
imshow('Black Canvas - Greayscale', image_gray, 4)
````

### Let's draw a line over our black square
`cv2.line(image, starting cordinates, ending cordinates, color, thikness)`

**Note:** Unlike many other OpenCV functions, this is an implace operation

````python
cv2.line(image, (0,0), (511,511), (255,0,0), 5)

imshow('Black Canvas With Diagonal Line', image, 5)
````

### Drawing rectangle

`cv2.line(image, starting vertex, opposite vertex, color, thikness)`



````python
# Create out black canvas again because it has a line in it
image= np.zeros((512,512,3), np.uint8)

# Positive thikness is line and negative thikness is filled
cv2.rectangle(image, (100,100), (200,300), (127,50,127), 10)
imshow('Black canvas with rectangle', image, 5)

````

### Drawing circle

`cv2.circle(image, center, radius, color, fill)`

````python
image= np.zeros((512,512,3), np.uint8)

cv2.circle(image, (250,250), 100, (50,127,50), -1)
imshow('Blacl canvas with circle', image, 5)
````

### **Polygons**

```cv2.polylines(image, points, Closed?, color, thickness)```

if Closed = True, we join the first and last points.

````python
image= np.zeros((512,512,3), np.uint8)

pts= np.array([[10,500],[50,100],[400,200],[150,250]], np.int32)

# Let's now reshape our points in form  required by polylines
pts= pts.reshape(-1,1,2)

cv2.polylines(image, [pts], True, (0,0,255), 3)

imshow('Black canvas with red polygon', image,5)
````

### **Adding text with cv2.putText**

```cv2.putText(image, 'Text to Display', bottom left starting point, Font, Font Size, Color, Thickness)```

**Available Fonts**
- FONT_HERSHEY_SIMPLEX, FONT_HERSHEY_PLAIN
- FONT_HERSHEY_DUPLEX,FONT_HERSHEY_COMPLEX 
- FONT_HERSHEY_TRIPLEX, FONT_HERSHEY_COMPLEX_SMALL
- FONT_HERSHEY_SCRIPT_SIMPLEX
- FONT_HERSHEY_SCRIPT_COMPLEX

````python
image= np.zeros((512,512,3), np.uint8)
string="Hello World!"
cv2.putText(image, string, (30,100), cv2.FONT_HERSHEY_SCRIPT_COMPLEX, 2, (20,200,30), 2)
imshow('Messing with some text', image, 5)
````
