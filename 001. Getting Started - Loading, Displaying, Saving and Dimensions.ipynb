### Dispale image

```python
import  matplotlib.pyplot as plt
```

```python
plt.imshow(image)
```

```python
plt.imshow(cv2.cvtColor(image,cv2.COLOR_RGB2BGR))
```

```python
def imshow(image=None,title=""):
    plt.imshow(cv2.cvtColor(image,cv2.COLOR_RGB2BGR))
    plt.title(title)
```

```python
flowers_image= cv2.imread("images/flowers.jpg")
```

```python
imshow(flowers_image,"flower store")
```

### Saving images
